Intro to Linear Regression f_wb = w*x[i]+b

```python
import matplotlib.pyplot as plt
import numpy as np
x_train = np.array([1.0,2.0,3.0])
y_train = np.array([300.0,500.0,700.0])
def calculate_house_price(x,w,b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
w = 200
b = 100
test_chp = calculate_house_price(x_train,w,b)
plt.title("Housing Price Prediction")
plt.plot(x_train,test_chp,color="blue",label="prediction")
plt.scatter(x_train,y_train,color="red",marker="x",s=100,label="training data")
plt.xlabel("Size of house(1000 sqft)")
plt.ylabel("price of house ($1000)")
plt.legend()
plt.tight_layout()
plt.show()
```
