Cost function basic Implementation.

model: f_wb = wx+b,     

parameters: w,b,  

cost func: J(w,b) = (1/2m)*summation of (f_wb_i - y_i)**2,    
        
goal: minimize of J(w,b) w,b.           

```python
# house price prediction
import matplotlib.pyplot as plt
import numpy as np

def calculate_cost(x,y,w,b):
    m = x.shape[0]
    cost = 0
    for i in range(m):
        f_wb = w*x[i]+b 
        cost_wb = (f_wb - y[i])**2
        cost += cost_wb 
    total_cost = (1/(2*m)) * cost
    return total_cost

x_train = np.array([1,2,3,4,5])
y_train = np.array([300,500,700,900,1110])
w = np.array([0,50,100,150,200,250,300,350,400])
cost_values = []
for value in w:
    cost_value = calculate_cost(x_train,y_train,value,100)
    cost_values.append(cost_value)
```

```python
plt.scatter(x_train,y_train,marker="x",c="r",label="Original Price")
plt.xlabel("House Size(1000 sqft)",family="Arial",fontsize=15)
plt.ylabel("House Price($1000)",family="Arial",fontsize=15)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(w,cost_values,marker="o",label="Prediction(w vs cost)")
plt.xlabel("W-weight",family="Arial",fontsize=15)
plt.ylabel("Cost",family="Arial",fontsize=15)
plt.tight_layout()
plt.legend()
plt.show()
```

```python

def predict_model(x,w,b):
    f_wb = w*x + b
    return f_wb
# exp x=3.5,w=200,b=100
x = 3.5
w = 200
b = 100
print(f"The House size {x*1000} sqft = price ${predict_model(x,w,b)}k")

# Final Visualization
plt.scatter(x_train,y_train,c="r",marker="x",label="Actual Price")
plt.plot(x_train,predict_model(x_train,w,b),label="Prediction")
plt.xlabel("House Size(1000 sqft)",family="Arial",fontsize=15)
plt.ylabel("House Price($1000)",family="Arial",fontsize=15)
plt.legend()
plt.show()
```
